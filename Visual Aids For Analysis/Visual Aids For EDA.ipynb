## 1. Line Chart


```python
# let's gerate file for this visualization

# let's import the libraries we will be using
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the dataset

dataset = pd.read_csv('stock.csv')
dataset.head(4)
```

```python
# let's check the dataset

dataset.dtypes
```

```python
# let's change the data type for date to datetime data type
dataset['Date'] = pd.to_datetime(dataset['Date'], format='%Y-%m-%d')
```

```python
# now 

dataset.dtypes
```

```python
#let's group the "Date" and get the mean

dataset_groupby_mean = dataset.groupby('Date').mean()
```

```python
# creating the plot

plt.figure(figsize= (14,10))
plt.title("Line Chat", size = 30)
plt.ylabel('Price', size = 20)
plt.xlabel('Date', size = 20)
plt.plot(dataset_groupby_mean)
plt.savefig('Linechat.png', bbox_inches='tight')

plt.show();
```

## 2. Bar Chart 

```python
# Here we will create a dataset to create the bar chart

import numpy as np
import calendar
import random

months = list(range(1,13))
sold_quantity = [round(random.uniform(100,200)) for x in range(1,13)]
```

```python
# plot the graph

figure, axis = plt.subplots(figsize = (14,10))
plt.xticks(months, calendar.month_name[1:13], rotation = 20)
plt.title("Bar Chat", size = 30)
plt.ylabel('Sold_quantity', size = 20)
plt.xlabel('Months', size = 20)
plot = axis.bar(months, sold_quantity)
for rectangle in plot:
    height = rectangle.get_height()
    axis.text(rectangle.get_x() + rectangle.get_width() /2., 1.002 * 
             height, '%d' % int(height), ha='center', va = 'bottom')
plt.savefig('barchat.png', bbox_inches='tight')
plt.show();
```

```python
months = list(range(1,13))
sold_quantity = [round(random.uniform(100,200)) for x in range(1,13)]

# creating the plot
figure, axis = plt.subplots(figsize = (14,10))

plt.yticks(months, calendar.month_name[1:13], rotation = 20)

plot = axis.barh(months, sold_quantity)
plt.title('Bar Chat', size=30)
plt.xlabel('Sold_quantity', size = 20)
plt.ylabel('Months', size = 20)

for rectangle in plot:
    width = rectangle.get_width()
    axis.text(width + 2.5, rectangle.get_y() + 0.38, '%d' % int(width), ha = 'center', va ='bottom')

plt.savefig('barhchat.png', bbox_inches='tight')
plt.show();
```

## 3. Scatter plot

```python
import seaborn as sns
sns.set()
```

```python
sleepDf = pd.read_csv('sleep_age.csv')
sleepDf.head(4)
```

```python

plt.figure(figsize = (14,10))
plt.scatter(x = sleepDf['age']/12., y = sleepDf['min_recommended'], label = 'min_recommended')
plt.scatter(x = sleepDf['age']/12., y = sleepDf['max_recommended'], label = 'max_recommended')
plt.xlabel('Age of person in years', size=20)
plt.ylabel('Total hours of sleep required', size = 20)
plt.title('Scatter plot: Sleep Recommendation per Age', size=30)
plt.legend()
plt.savefig('sleepscatterplot.png', bbox_inches='tight')
plt.show();
```

```python
plt.figure(figsize = (14,10))
plt.plot(sleepDf['age']/12., sleepDf['min_recommended'], 'g--', label = 'min_recommended' )
plt.plot(sleepDf['age']/12., sleepDf['max_recommended'], 'r--', label = 'max_recommended')
plt.title('Scatter plot: Sleep Recommendation per Age', size=30)
plt.xlabel('Age of person in Years')
plt.ylabel('Total hours of sleep required')
plt.legend()
plt.savefig('sleepscatterplot2.png', bbox_inches='tight')
plt.show();
```

```python
# using seaborn to load the dataset

plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['figure.dpi'] = 150

df = sns.load_dataset('iris')
df['species'] = df['species'].map({'setosa':0, 'versicolor':1, 'virginica':2})

plt.scatter(x = df['sepal_length'], y = df['sepal_width'], c=df.species)
plt.title('Scatter plot: Iris Dataset ')

plt.xlabel('Sepal length')
plt.ylabel('Petal length')

plt.show();
```
