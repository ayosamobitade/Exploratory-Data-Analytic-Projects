### Line Chart

```python
# let's gerate file for this visualization

# let's import the libraries we will be using
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the dataset

dataset = pd.read_csv('stock.csv')
dataset.head(4)
```

```python
dataset_groupby_mean = dataset.groupby(by = 'Date').mean()
```

```python
plt.figure(figsize= (14,10))
plt.plot(dataset_groupby_mean)
plt.show();
```
