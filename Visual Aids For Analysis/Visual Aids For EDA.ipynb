## 1. Line Chart


```python
# let's gerate file for this visualization

# let's import the libraries we will be using
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the dataset

dataset = pd.read_csv('stock.csv')
dataset.head(4)
```

```python
# let's check the dataset

dataset.dtypes
```

```python
# let's change the data type for date to datetime data type
dataset['Date'] = pd.to_datetime(dataset['Date'], format='%Y-%m-%d')
```

```python
# now 

dataset.dtypes
```

```python
#let's group the "Date" and get the mean

dataset_groupby_mean = dataset.groupby('Date').mean()
```

```python
# creating the plot

plt.figure(figsize= (14,10))
plt.title("Line Chat", size = 30)
plt.xlabel('Price', size = 20)
plt.ylabel('Date', size = 20)
plt.plot(dataset_groupby_mean)

plt.show();
```

## 2. Bar Chart 

```python
# Here we will create a dataset to create the bar chart

import numpy as np
import calendar
import random

months = list(range(1,13))
sold_quantity = [round(random.uniform(100,200)) for x in range(1,13)]
```

```python
# plot the graph

figure, axis = plt.subplots(figsize = (14,10))
plt.xticks(months, calendar.month_name[1:13],rotation = 20)
plt.title("Bar Chat", size = 30)
plt.xlabel('Sold_quantity', size = 20)
plt.ylabel('Months', size = 20)
plot = axis.bar(months, sold_quantity)
for rectangle in plot:
    height = rectangle.get_height()
    axis.text(rectangle.get_x() + rectangle.get_width() /2., 1.002 * 
             height, '%d' % int(height), ha='center', va = 'bottom')
plt.show();
```
